```python
# Question 1

import csv

total_revenue = 0

with open("sales_data.csv","r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        quantity = int(row[5])
        price = int(row[6])
        revenue = quantity * price
        total_revenue += revenue

print("Total Revenue:", total_revenue)
```

```python
# Question 4

import csv

category_revenue = {}

with open("sales_data.csv","r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        category = row["Category"]
        quantity = int(row["Quantity"])
        price = float(row["Price"])
        
        revenue = quantity * price
        
        if category not in category_revenue:
            category_revenue[category] =  0
            
        category_revenue[category] += revenue

print("Total Revenue by Category:")
for category in category_revenue:
    print(category, ":", category_revenue[category])
    
            
```

```python
# Question 5

import csv

product_revenue = {}

with open("sales_data.csv","r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        product = row["Product"]
        quantity = int(row["Quantity"])
        price = float(row["Price"])
        
        revenue = quantity * price
        
        if product not in product_revenue:
            product_revenue[product] = 0
            
        product_revenue[product] += revenue
        
print("Total Revenue by Product:")    
for product in product_revenue:
    print(product,product_revenue[product])   
```

```python
# Question 6

import csv

product_quantity = {}

with open("sales_data.csv","r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        product = row[3]
        quantity = int(row[5])

        if product in product_quantity:
            product_quantity[product] += quantity
        else:
            product_quantity[product] = quantity

max_product = max(product_quantity, key=product_quantity.get)

print("Product with highest quantity sold:", max_product)
print("Total Quantity:", product_quantity[max_product])
```

```python
# Question 7

import csv

customer_spending = {}

with open("sales_data.csv","r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        customer = row[2]
        quantity = int(row[5])
        price = int(row[6])
        revenue = quantity * price

        if customer in customer_spending:
            customer_spending[customer] += revenue
        else:
            customer_spending[customer] = revenue

max_customer = max(customer_spending, key=customer_spending.get)

print("Customer who spent the most:", max_customer)
print("Total Spending:", customer_spending[max_customer])
```

```python
# Question 8

import csv

date_revenue = {}

with open("sales_data.csv","r") as file:
    reader = csv.reader(file)
    next(reader)   

    for row in reader:
        date = row[1]
        quantity = int(row[5])
        price = int(row[6])

        revenue = quantity * price

        if date in date_revenue:
            date_revenue[date] += revenue
        else:
            date_revenue[date] = revenue

print("Revenue generated for each date:")

for date, revenue in date_revenue.items():
    print(date, ": ", revenue)
```

```python
# Question 9

import csv

start_date = input("Enter start date (YYYY-MM-DD): ")
end_date = input("Enter end date (YYYY-MM-DD): ")

total_revenue = 0

with open("sales_data.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        date = row["Date"]
        quantity = int(row["Quantity"])
        price = float(row["Price"])

        revenue = quantity * price

        if start_date <= date <= end_date:
            total_revenue += revenue

print("Total Revenue between", start_date, "and", end_date, ": ₹", total_revenue)
```

```python
# Question 10

import csv

with open("sales_data.csv","r") as file, open("electronics_orders.csv","w",newline="") as newfile:
    
    reader = csv.reader(file)
    writer = csv.writer(newfile)

    header = next(reader)
    writer.writerow(header)

    for row in reader:
        if row[4] == "Electronics":
            writer.writerow(row)

print("Electronics orders exported successfully")
```
